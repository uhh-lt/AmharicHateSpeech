## step- 2 Removing unnecessary punctuation and remove url and gmail Charactor

```python
# Output:

import re



def removing_url_gmail_date(string_list):
    result_list = []
    for string in string_list:
        if isinstance(string, str):   # check if string is a string type
            string = re.sub(r'http\S+', '', string) # used to remove url
            string = re.sub('https?://\S+|www\.\S+','',string)
            string = re.sub('([\w\.-]+)@([\w\.-]+)','',string) # used to remove gmail 
        result_list.append(string)
    return result_list

dataset['text'] = removing_url_gmail_date(dataset['text'].tolist())
def removing_uncecessary_charactor(string_list):
    result_list = []
    for string in string_list:
        if isinstance(string, str):   # check if string is a string type
            string = string.strip() # removing the uncessary word whitespaces
            string = re.sub('፤', '', string)
            string = re.sub("(\\d)+", '', string) # remove any digit 
            string = re.sub('፤!.፤፤"@#$%^«»/&*()…;"”"”›’‘":,.‹/<>?\|`=+፡።፤;፦፥፧፨፠፣', '', string)  # remove punctuation 
            string = re.sub('፠፡።፣፤፥፦፧፨“”‘’…‹‹››·•👉😂👇❗', '', string)
            string = re.sub('[a-zA-Z]', '', string) # Remove English alphabets
            string = re.sub('፩፪፫፬፭፮፯፰፱፲፳፴፵፶፷፸፹፺፻፼', '', string)
        else:
            print(string)
        result_list.append(string)
    return result_list

dataset['text'] = removing_uncecessary_charactor(dataset['text'].tolist())

```

```python
# list amharic stop words
# read the stop words file
with open('AmhStopWord.txt', 'r', encoding='utf-8') as f:
    stop_words = [line.strip() for line in f]
# define a function to remove stop words from a sentence
def remove_stopwords(sentence):
    sentence = re.sub(r'[^\w\s]', '', sentence) # remove punctuation
    words = sentence.split()
    words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(words)
# remove stop words from text column
dataset['text'] = dataset['text'].apply(remove_stopwords)
```
